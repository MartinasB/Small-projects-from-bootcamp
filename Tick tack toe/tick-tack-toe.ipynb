### TICK TACK TOE

Check it out and have fun as I did writing it and testing.

```python
import random
from IPython.display import clear_output
import time

marks = ["O","X"]
field = ['-'] * 9
win_seq = ((1, 2, 3), (4, 5 ,6), (7, 8, 9), (1, 4, 7), (2, 5, 8), (3, 6, 9), (1, 5, 9), (3, 5, 7))
win = False
no_moves = False
a_or_b = ""

def f_print_field(field): # prints field after chosing its indexing type
    clear_output(True)
    time.sleep(1)
    if a_or_b.lower() == "a":
        print(f'{field[6]} {field[7]} {field[8]}\n{field[3]} {field[4]} {field[5]}\n{field[0]} {field[1]} {field[2]}')
    elif a_or_b.lower() == "b":
        print(f'{field[0]} {field[1]} {field[2]}\n{field[3]} {field[4]} {field[5]}\n{field[6]} {field[7]} {field[8]}')
    else:
        print('Chose A or B field')

def f_insert_mark(index, mark, field): # replaces player's field list with chosen index
    index -= 1
    field.pop(index)
    field.insert(index, mark)

def f_check_win(win_seq, mark, field): # checks if X or O player is in winning sequence
    m_list = []
    i = 0
    while i < len(field): # creates list of marks index
        if field[i] == mark:
            m_list.append(i+1)
        i += 1
    i = 0
    do_we_have_a_winner = 0
    while i < len(win_seq) and do_we_have_a_winner == 0: # check if created list contains winning numbers sequence
        if str(win_seq[i][0]) in str(m_list) and str(win_seq[i][1]) in str(m_list) and str(win_seq[i][2]) in str(m_list):
            i += 1
            do_we_have_a_winner += 1
        else:
            i += 1
    if do_we_have_a_winner >= 1:
        return True
    else:
        return False
        
def f_check_if_moves_available(list): # checks if there are moves available
    if "-" in list:
        return False
    else:
        return True

def f_check_if_cell_is_empty(index, list): # checks if next move is possible (field's cell is empty)
    index -= 1
    if list[index] == "-":
        return True
    else:
        return False

print(f"""
TICK TACK TOE GAME
field indexing:
~~~~~~~~~~~~~
  A   or  B
7 8 9 | 1 2 3
4 5 6 | 4 5 6
1 2 3 | 7 8 9
~~~~~~~~~~~~~""")

time.sleep(1)
a_or_b = str(input("chose field indexing type: "))

player = random.choice(marks)

while a_or_b.lower() != 'a' and a_or_b.lower() != 'b':
    a_or_b = str(input("chose between A and B"))

f_print_field(field)
    
while win != True or no_moves != True:

    index = int(input(f'cell id for "{player}" mark: '))
    
    while index not in range(1,10):
        print("index out of range, chose number between 1 and 9")
        index = int(input(f'cell id for "{player}" mark: '))

    while f_check_if_cell_is_empty(index, field) == False:
        print("cell taken, chose onother one")
        index = int(input(f'cell id for "{player}" mark: '))
        
    f_insert_mark(index, player, field)
    if f_check_win(win_seq, player, field) == True:
        f_print_field(field)
        print(f'"{player}" won!')
        break
    if f_check_if_moves_available(field) == True:
        f_print_field(field)
        print("no more moves")
        break
   
    f_print_field(field)

    player = marks[(marks.index(player))-1] # switch player / mark
```
